```python
import folium
import pandas as pd
import re
```

```python
geo_json='https://raw.githubusercontent.com/southkorea/seoul-maps/master/kostat/2013/json/seoul_municipalities_geo_simple.json'
```

```python
df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/코로나관련/코로나 데이터 전처리/코로나(최신구별).csv')
df.head()
```

```python
df['연도'] = df['확진일'].str.split('-', expand=True)[0]
df['월'] = df['확진일'].str.split('-', expand=True)[1]
df
```

```python
#확진일에 일을 제외한 연도 월만 나오는 컬럼을 만든다
df['연도_월'] = df['확진일'].apply(lambda some_str: some_str[:-3])
df
```

```python
#2020 data
df_2020 = df.loc[df['확진일']<='2020-12-31', :]
df_2020.tail()
```

```python
#20201 data
df_2021 = df.loc[df['확진일']>='2021-01-01', :]
df_2021.head()
```

```python
# 전체 2020상에서 변화폭이 커지는 시점인 7월 이전과 이후로 나눔
df_2020_ss = df_2020.loc[df['확진일']<='2020-06-30', :]
df_2020_ss
```

```python
df_2020_fw = df_2020.loc[df['확진일']>='2020-07-01', :]
df_2020_fw
```

```python
# 각 기간 지역별 확진자수 합을 구한다.
df_2020_ss_sum = df_2020_ss[['지역','확진자수']].groupby('지역').sum()
df_2020_ss_sum = df_2020_ss_sum.reset_index()
df_2020_ss_sum.info()
```

```python
df_2020_fw_sum = df_2020_fw[['지역','확진자수']].groupby('지역').sum()
df_2020_fw_sum = df_2020_fw_sum.reset_index()
```

```python
df_2021_sum = df_2021[['지역','확진자수']].groupby('지역').sum()
df_2021_sum = df_2021_sum.reset_index()
```

```python
# df_2020_ss_sum[['지역','확진자수']]
```

```python
# 컬럼 명을 바꿔준다.
df_2020_ss_sum.columns = ['name','values']
# df_2020_ss_sum.head()

df_2020_ss_sum = df_2020_ss_sum.sort_values(by='name')
```

```python
df_2020_fw_sum.columns = ['name','values']

df_2020_fw_sum = df_2020_fw_sum.sort_values(by='name')
```

```python
df_2021_sum.columns = ['name','values']

df_2021_sum = df_2020_fw_sum.sort_values(by='name')
```

```python
m = folium.Map(
    location=[37.566345,126.977893],
    tiles='Stamen Terrain'
)

folium.Choropleth(
    geo_data=geo_json,
    name='choropleth',
    data=df_2020_ss_sum,
    columns=['name','values'],
    key_on='feature.properties.name',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.2
).add_to(m)

m
```
