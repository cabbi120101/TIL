```python
#csv 파일들 디렉토리 위치
input_file = r'/content/drive/MyDrive/Colab Notebooks/코로나관련/대중교통관련/지하철 데이터'

output_file = r'/content/data/sub_all_all.csv'

allfile_list=glob.glob(os.path.join(input_file, 'CARD_*.csv'))
print(allfile_list)
```

```python
allData = []
for thisfile in allfile_list:
    
    try:
        df = pd.read_csv(thisfile,encoding='UTF-8')
        allData.append(df)
        display(df.head())


    except:
        try:
            df = pd.red_csv(thisfile, encoding='cp949')
            allData.append(df)
            display(df.head())
        except:
            try:
                df = pd.red_csv(thisfile, encoding='latin-1')
                allData.append(df)
                display(df.head())
            except:
                try:
                    df = pd.red_csv(thisfile, encoding='uft-16')
                    allData.append(df)
                    display(df.head())
                except:
                    print(thisfile,'예외')
                
        
               
dataCombine = pd.concat(allData, axis=0, ignore_index=True)

dataCombine.to_csv(output_file, index=False)

# big_df = pd.merge(df, bus_df, how="left", on="역명")
# big_df.head()
```
