```python
katok = ['다현','정연','쯔위','사나','지효']
```

```python
def insert_data(position, friend) : #삽입할위치, 친구
    katok.append(None)
    kLen = len(katok)

    for i in range(kLen-1, position, -1) :
        katok[i] = katok[i-1]
        katok[i-1] = None


    katok[position] = friend
```

```python
insert_data(2, '솔라')
print(katok)
insert_data(6, '문별')
print(katok)
```
