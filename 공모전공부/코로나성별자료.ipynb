```python
from urllib.request import urlopen, Request
from urllib.parse import urlencode, unquote, quote_plus
import urllib
import requests
import json
import pandas as pd
import xmltodict # 결과가 xml 형식으로 반환된다. 이것을 dict 로 바꿔주는 라이브러리다
```

```python
#성휴 코드
key = 'DECODING KEY'

num_of_rows = 10 # 한페이지에 포함된 결과 수 
page_no = 1 #페이지 번호
url = 'http://openapi.data.go.kr/openapi/service/rest/Covid19/getCovid19GenAgeCaseInfJson'
start_Date = 20200101
end_Date = 20210705

req_parameter = '?' + urlencode({
    'serviceKey':  key,
    'numOfRows' : num_of_rows,
    'pageNo' : page_no,
    'startCreateDt' : start_Date,
    'endCreateDt' : end_Date,
})

r = requests.get(url + req_parameter)
response_content = json.loads(json.dumps(xmltodict.parse(r.text)))
response_content['response']['body']['items']['item']
```

```python
item = response_content['response']['body']['items']['item']
df = pd.DataFrame(item)
df.to_csv('/content/drive/코로나성별.csv', index = False)
```
