```python
import pandas as pd
```

```python
#저장한 csv파일 불러오기(encoding 할 필요 없음)
data_file = '/content/drive/코로나(시.도20200301~20210705) .csv'
df = pd.read_csv(data_file)
df
```

```python
#결측값 확인
df.isnull()
```

```python
#결측치 개수 확인
df.isnull().sum()
```

```python
#변수 df에서 필요없다고 생각하는 열을 제거함
df =df.drop(columns=['createDt','deathCnt','gubunCn','gubunEn','isolClearCnt','isolIngCnt','overFlowCnt','updateDt','seq'])
df
```

```python
# df에서 결측치가 있는 행 모두 제거함
# 결측치를 0으로 대체 하고 싶은경우
# df.fillna(0) 사용
# 열의 다음값을 넣고 싶은 경우 (method='bfill')
# 열의 이전 값을 넣을 경우 (method='ffill')
df.dropna()
```

```python
#만든 데이터 확
df.info()
```

```python
#열이름을 보기 편하게 변경
df.rename(columns={'defCnt':'확진자수','gubun':'지역','incDec':'전일대비증감수','localOccCnt':'지역발생수','qurRate':'10만명당발생률','stdDay':'기준일'}, inplace=True)
```

```python
# 변경 확
df.info()
```

```python
# 다시 결측값 있는지 확인
df.isnull().sum()
```

```python
# 결측값 있는 index 삭제
df = df.dropna()
df
```

```python
#다시 csv로 저장 

df = pd.DataFrame(df)
df.to_csv('/content/drive/코로나(시도 정리).csv', index = False)
```
