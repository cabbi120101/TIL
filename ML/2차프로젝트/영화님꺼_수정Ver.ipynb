```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
rent = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/멀티캠/머신러닝/2차프로젝트/홍영화님/rent3.csv',index_col = 0)
```

```python
rent.head(30)
```

```python
rent_기타 = rent[rent['광역상권']=='기타지역']
rent_기타.head()
```

```python
rent_이외 = rent[rent['광역상권']!='기타지역']
rent_이외.head()
```

```python
# 기타지역을 평균으로 묶음
rent_기타mean = rent_기타.groupby(['연도','분기','광역상권'])[['임대료','공실률','투자수익률','소득수익률','자본수익률']].mean().reset_index()
rent_기타mean.head()
```

```python
rent_기타mean = rent_기타mean.rename(columns={"광역상권": "하위상권"})
```

```python
rent_all = pd.concat([rent_이외, rent_기타mean], axis=0)
rent_all.tail()
```

```python
rent_all = rent_all.drop(['광역상권'], axis=1)
```

```python
rent_all = rent_all[rent_all['연도'] != 2021]
rent_all
```

```python
rent_all_oh = pd.get_dummies(rent_all, columns=["하위상권"])
rent_all_oh.head()
```

```python
train = rent_all_oh[rent_all_oh['연도'] != 2020]
train.tail()
```

```python
test = rent_all_oh[rent_all_oh['연도'] == 2020]
test.head()
```
